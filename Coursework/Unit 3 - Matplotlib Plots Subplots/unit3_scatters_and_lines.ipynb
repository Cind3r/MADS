The following cell provides some data.

```python
import random

num_points = 10
x1 = range(num_points)
y1 = [random.randint(0, 10) for _ in range(num_points)]
z1 = [random.randint(5, 20) for _ in range(num_points)]

x2 = range(num_points)
y2 = [random.randint(0, 10) for _ in range(num_points)]
z2 = [random.randint(5, 20) for _ in range(num_points)]
```

In the cell below, correctly import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

Create a new figure with one subplot

```python
fig = plt.figure(figsize=(8, 6))
fig.add_subplot(111)
```

Plot the `x1` and `y1` variables as a line plot with their data on their corresponding axes.

```python
plt.plot(x1, y1)
plt.xlabel('X')
plt.ylabel('Y')
```

Add the plot of `x2` and `y2` to the same plot.

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1', 'y2'])
```

Change the colors of the lines to `lightblue` (`x1/y1`) and `darkblue` (`x2/y2`).

```python
plt.plot(x1, y1, color='lightblue')
plt.plot(x2, y2, color='darkblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1', 'y2'])
```

Label the x-axis `X` and y-axis `Y`

```python
plt.plot(x1, y1, color='lightblue')
plt.plot(x2, y2, color='darkblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1', 'y2'])
```

Title the plot `Random data`

```python
plt.plot(x1, y1, color='lightblue')
plt.plot(x2, y2, color='darkblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random data')
plt.legend(['y1', 'y2'])
```

Now switch the plot to a scatter plot.

```python
plt.scatter(x1, y1, c='lightblue')
plt.scatter(x2, y2, c='darkblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random data')
plt.legend(['y1', 'y2'])
```

Finally, make the marker sizes for `x1/y1` correspond to `z1` and `x2/y2` correspond to `z2`

```python
plt.scatter(x1, y1, s=z1, c='lightblue')
plt.scatter(x2, y2, s=z2, c='darkblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random data')
plt.legend(['y1', 'y2'])
```
